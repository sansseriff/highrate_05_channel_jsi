```python
data_set = {}
chA_list = [35,36,37,38,39,40,41,42]
chB_list = [52,53,54,55,56,57,58,59]
for chA in chA_list:
    data_set[chA] = {}
    path = "../data/"
    for chB in chB_list:
        try:
            with open(os.path.join(path, f"ent_data_ch{chA}_{chB}.json"), 'r') as file:
                data = json.load(file)
                data_set[chA][chB] = data
        except:
#             traceback.print_exc()
            data = {}
            data_set[chA][chB] = data
```
