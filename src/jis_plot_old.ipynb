```python
import json
import matplotlib.pyplot as plt
import numpy as np
chB_list = [34,35,36,37,38,39,40,41,42]
chA_list = [52,53,54,55,56,57,58,59]
from phd import viz

colors, swatches = viz.PHD
```

```python
data_set = {}

for chA in chA_list:
    data_set[chA] = {}
    for chB in chB_list:
        try:
            with open(f"ent_data_ch{chA}_{chB}.json", 'r') as file:
                data = json.load(file)
                data_set[chA][chB] = data
        except:
            data = {}
            data_set[chA][chB] = data
```

```python
st = data_set[52][41]

plt.plot(st["hist_data_index"],st["hist_data"])
sum(st["hist_data"][750:1250])
start = 750//2
end = 1250//2
sum(st["hist_data"][start:end])


for chA in chA_list:
    for chB in chB_list:
        try:
            st = data_set[chA][chB]
            data_set[chA][chB]["coinc"] = sum(st["hist_data"][start:end])
        except:
            data_set[chA][chB]["coinc"] = 0
```

```python
Set = []
Set_names = []
for i, chA in enumerate(chA_list):
    Set.append([])
    
    for j, chB in enumerate(chB_list):
        try: 
            coinc = data_set[chA][chB]["coinc"]
            Set[i].append(coinc)
        except:
            Set[i].append(0)
            
```
